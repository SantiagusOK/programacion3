# **Interfaces**

imagen.png

 Las interfaces son una forma de definir un contrato que las clases deben seguir.
 En Python, no existen interfaces como un tipo específico (como en Java o C#), pero se pueden crear interfaces utilizando clases abstractas del módulo abc (Abstract Base Classes o Clases Base Abstractas).

**¿Para qué sirven las interfaces?**

  **1.  Definen un contrato:** Especifican un conjunto de métodos que una clase debe implementar.

  **2. Promueven el polimorfismo:** Permiten que diferentes clases implementen la misma interfaz de diferentes maneras, facilitando el uso de objetos de manera intercambiable.

  **3. Fomentan el diseño desacoplado:** Ayudan a reducir las dependencias entre componentes, facilitando el mantenimiento y la evolución del código.

**Ejemplo de interfaces en Python**

Supongamos que estamos desarrollando una aplicación de procesamiento de pagos.

Queremos permitir diferentes métodos de pago como tarjetas de crédito, PayPal y criptomonedas.

Vamos a definir una interfaz para los métodos de pago y luego implementaremos esta interfaz en diferentes clases.

**Paso 1: Definir la interfaz de método de pago**

```python
from abc import ABC, abstractmethod

class MetodoPago(ABC):

    @abstractmethod
    def procesar_pago(self, cantidad):
        pass
```

**Paso 2: Implementar la interfaz en diferentes clases**

```python
class TarjetaCredito(MetodoPago):

    def procesar_pago(self, cantidad):
        print(f"Procesando un pago de {cantidad} usando Tarjeta de Crédito")

class PayPal(MetodoPago):

    def procesar_pago(self, cantidad):
        print(f"Procesando un pago de {cantidad} usando PayPal")

class Criptomoneda(MetodoPago):

    def procesar_pago(self, cantidad):
        print(f"Procesando un pago de {cantidad} usando Criptomoneda")
```

**Paso 3: Usar las clases que implementan la interfaz**

```python
# Función que acepta cualquier objeto que implemente la interfaz MetodoPago
def realizar_pago(metodo_pago: MetodoPago, cantidad):
    metodo_pago.procesar_pago(cantidad)

# Crear instancias de diferentes métodos de pago
tarjeta = TarjetaCredito()
paypal = PayPal()
crypto = Criptomoneda()

# Realizar pagos usando diferentes métodos
realizar_pago(tarjeta, 100)
realizar_pago(paypal, 200)
realizar_pago(crypto, 300)
```

**Explicación del código**

**Definición de la interfaz MetodoPago:**

> MetodoPago es una clase abstracta que define un método abstracto procesar_pago.

> Cualquier clase que implemente esta interfaz debe proporcionar una implementación de este método.

**Implementación de la interfaz en diferentes clases:**

TarjetaCredito, PayPal, y Criptomoneda son clases concretas que implementan la interfaz MetodoPago. Cada una proporciona su propia versión del método procesar_pago.

**Uso de las clases que implementan la interfaz:**

La función realizar_pago acepta cualquier objeto que implemente la interfaz MetodoPago. Esto permite que la función trabaje con diferentes tipos de métodos de pago de manera intercambiable.

Se crean instancias de TarjetaCredito, PayPal y Criptomoneda, y se pasan a la función realizar_pago junto con una cantidad. Cada instancia procesa el pago de acuerdo con su propia implementación del método procesar_pago.

**Beneficios de usar interfaces**

1. Flexibilidad: Permite cambiar o agregar nuevos métodos de pago sin modificar el código que usa estos métodos.

2. Reutilización: El mismo código puede trabajar con diferentes implementaciones de métodos de pago.

3. Mantenimiento: Facilita el mantenimiento y la evolución del código, ya que las implementaciones están desacopladas de las interfaces.

Este ejemplo demuestra cómo las interfaces (implementadas mediante clases abstractas) pueden ser utilizadas para definir contratos en Python y cómo pueden proporcionar un diseño más flexible y desacoplado.
