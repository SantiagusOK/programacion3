# **Interfaces**

imagen.png

 Las interfaces son una forma de definir un contrato que las clases deben seguir.
 En Python, no existen interfaces como un tipo especfico (como en Java o C#), pero se pueden crear interfaces utilizando clases abstractas del mdulo abc (Abstract Base Classes o Clases Base Abstractas).

**Para qu sirven las interfaces?**

  **1.  Definen un contrato:** Especifican un conjunto de mtodos que una clase debe implementar.

  **2. Promueven el polimorfismo:** Permiten que diferentes clases implementen la misma interfaz de diferentes maneras, facilitando el uso de objetos de manera intercambiable.

  **3. Fomentan el diseo desacoplado:** Ayudan a reducir las dependencias entre componentes, facilitando el mantenimiento y la evolucin del cdigo.

**Ejemplo de interfaces en Python**

Supongamos que estamos desarrollando una aplicacin de procesamiento de pagos.

Queremos permitir diferentes mtodos de pago como tarjetas de crdito, PayPal y criptomonedas.

Vamos a definir una interfaz para los mtodos de pago y luego implementaremos esta interfaz en diferentes clases.

**Paso 1: Definir la interfaz de mtodo de pago**

```python
from abc import ABC, abstractmethod

class MetodoPago(ABC):

    @abstractmethod
    def procesar_pago(self, cantidad):
        pass
```

**Paso 2: Implementar la interfaz en diferentes clases**

```python
class TarjetaCredito(MetodoPago):

    def procesar_pago(self, cantidad):
        print(f"Procesando un pago de {cantidad} usando Tarjeta de Crdito")

class PayPal(MetodoPago):

    def procesar_pago(self, cantidad):
        print(f"Procesando un pago de {cantidad} usando PayPal")

class Criptomoneda(MetodoPago):

    def procesar_pago(self, cantidad):
        print(f"Procesando un pago de {cantidad} usando Criptomoneda")
```

**Paso 3: Usar las clases que implementan la interfaz**

```python
# Funcin que acepta cualquier objeto que implemente la interfaz MetodoPago
def realizar_pago(metodo_pago: MetodoPago, cantidad):
    metodo_pago.procesar_pago(cantidad)

# Crear instancias de diferentes mtodos de pago
tarjeta = TarjetaCredito()
paypal = PayPal()
crypto = Criptomoneda()

# Realizar pagos usando diferentes mtodos
realizar_pago(tarjeta, 100)
realizar_pago(paypal, 200)
realizar_pago(crypto, 300)
```

**Explicacin del cdigo**

**Definicin de la interfaz MetodoPago:**

> MetodoPago es una clase abstracta que define un mtodo abstracto procesar_pago.

> Cualquier clase que implemente esta interfaz debe proporcionar una implementacin de este mtodo.

**Implementacin de la interfaz en diferentes clases:**

TarjetaCredito, PayPal, y Criptomoneda son clases concretas que implementan la interfaz MetodoPago. Cada una proporciona su propia versin del mtodo procesar_pago.

**Uso de las clases que implementan la interfaz:**

La funcin realizar_pago acepta cualquier objeto que implemente la interfaz MetodoPago. Esto permite que la funcin trabaje con diferentes tipos de mtodos de pago de manera intercambiable.

Se crean instancias de TarjetaCredito, PayPal y Criptomoneda, y se pasan a la funcin realizar_pago junto con una cantidad. Cada instancia procesa el pago de acuerdo con su propia implementacin del mtodo procesar_pago.

**Beneficios de usar interfaces**

1. Flexibilidad: Permite cambiar o agregar nuevos mtodos de pago sin modificar el cdigo que usa estos mtodos.

2. Reutilizacin: El mismo cdigo puede trabajar con diferentes implementaciones de mtodos de pago.

3. Mantenimiento: Facilita el mantenimiento y la evolucin del cdigo, ya que las implementaciones estn desacopladas de las interfaces.

Este ejemplo demuestra cmo las interfaces (implementadas mediante clases abstractas) pueden ser utilizadas para definir contratos en Python y cmo pueden proporcionar un diseo ms flexible y desacoplado.
